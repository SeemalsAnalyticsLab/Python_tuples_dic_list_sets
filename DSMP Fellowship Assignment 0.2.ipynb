## **Overview of the Attendance Management Assignment**

This assignment focuses on the implementation of basic data structures in Python to manage student attendance effectively. It explores the use of lists to record attendance, sets to identify unique attendees, tuples to store immutable records, and dictionaries to count occurrences of each student's name. Through practical coding exercises, we will create an attendance list, identify unique names, store data in a permanent format, and count attendance accurately. By the end of this assignment, you will gain a deeper understanding of how to utilize different data structures for organizing and analyzing data in Python, laying a solid foundation for more complex programming tasks.

**Question 1: Create and Count Attendance List**

Task:

1. Create a list attendance_list with names, some repeated.

2. Print the list.

3. Count how many names (including duplicates) are in the list and print the total.

```python
# Step 1: Creating a list of student names for attendance check:

Attendance_list = ["John", "Emma", "John", "Ava", "Sophia", "Emma"]

# Step 2: Printing the set to see the unique student names:
print("Attendance List:", Attendance_list)

# Step 3: Counting how many names (including duplicates) are in the list:

total_names = len(Attendance_list)
print("Total number of names (including duplicates):", total_names)
```

**Explanation:**

A list attendance_list is created with several student names.
The list includes duplicate names, meaning some students attended more than once.

The len() function is used to count all elements, including duplicates.



**Question 2: Unique Attendance using a Set**

Task:

1. Convert the attendance_list into a set to find unique attendees (sets remove duplicates).

2. Print the set to display unique student names.

3. Count and print the total number of unique attendees.



```python
# Step 1: Defining the attendance list with duplicates
attendance_list = ["John", "Emma", "John", "Ava", "Sophia", "Emma"]

# Step 2: Converting the list to a set to remove duplicates
Unique_Attendance = set(attendance_list)

# Step 3: Printing the set to see the unique student names
print("Unique Attendance Set:", Unique_Attendance)

# Step 4: Counting and printing the total number of unique students
total_names = len(Unique_Attendance)
print("Total Unique Attendance:", total_names)
```

**Explanation:**

Sets in Python automatically remove duplicates. So converting a list into a set helps us find unique attendees.

len() function is used again, but now it counts only the unique elements (names).

The set will remove any duplicate names like "John" and "Emma", ensuring that each student's name appears only once.


**Question 3: Storing Attendance Permanently with a Tuple**

Task:

1. Store the attendance list in a tuple, which is immutable (cannot be changed).

2. Print the tuple to display the names.

```python
# Step 1: Creating the tuple called attendance_tuple
attendance_tuple = ("John", "Emma", "Ava", "Sophia")

# Step 2: Printing the tuple to display the names
print("Attendance Tuple:", attendance_tuple)
```

**Explanation**:

A tuple is similar to a list, but it’s immutable. Once created, its elements cannot be changed.

This is useful when you want to store data that shouldn’t be modified later, such as historical attendance records.

The names are stored in attendance_tuple, and the print() function is used to display them.

**Question 4: Count Attendance using a Dictionary**

Task:

1. Using the attendance_list, create a dictionary where each student's name is a key, and the value is the count of how many times they attended.

2. Print the dictionary to see each student's attendance count.
Solution:

```python
# Step 1: Using the attendance_list from Question 1
attendance_list = ["John", "Emma", "John", "Ava", "Sophia", "Emma"]

# Step 2: Creating an empty dictionary to store the attendance count
attendance_count = {}

# Step 3: Counting the number of times each student attended
for student in attendance_list:
    if student in attendance_count:
        attendance_count[student] += 1
    else:
        attendance_count[student] = 1

# Step 4: Printing the dictionary to see each student's attendance count
print("Attendance Count Dictionary:", attendance_count)
```

**Explanation:**

A dictionary is used because it allows us to store key-value pairs. The keys are the student names, and the values are the number of times they attended.
The for loop iterates over each student in attendance_list. If the student's name already exists in attendance_count, we increment the count; otherwise, we add the name to the dictionary with an initial count of 1.
Finally, we print the dictionary to see how many times each student attended the class.


 ## **Summary of Concepts:**

**List:** Used for storing items (attendance list), can contain duplicates.

**Set:** Used for storing unique items (unique attendance), removes duplicates automatically.

**Tuple:** Used for storing immutable data, cannot be changed after creation.

**Dictionary:** Used to store key-value pairs (student names as keys and their attendance count as values). Perfect for counting occurrences of items in a list.

This approach demonstrates fundamental Python data structures (lists, sets, tuples, dictionaries) and operations like counting, looping, and displaying data effectively!
